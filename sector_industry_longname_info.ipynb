```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.style as style
import plotly as py
from plotly.offline import init_notebook_mode, iplot
import chart_studio.tools as tls
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os
import glob
import linearmodels
import yfinance as yf
import json

base = "C:/Users/Linchen Zhang/Desktop/fall2020/linchen_robintrack/"
indir = "C:/Users/Linchen Zhang/Desktop/fall2020/linchen_robintrack/robintrack-popularity-history/tmp/popularity_export"
clean = "C:/Users/Linchen Zhang/Desktop/fall2020/linchen_robintrack/cleanPopularity/"
```

```python
df = pd.read_csv(base + "users_prices.csv")
df.head()
```

```python
stocks = df['Symbol'].unique()

ticker_sector_dict = {}
ticker_industry_dict = {}
ticker_longname_dict = {}
```

```python
count = 0
for s in stocks:
    count += 1
    if count%100 == 0:
        print(count)
    try:
        ticker = yf.Ticker(s)
        js = json.dumps(ticker.info)
        fp = open(base+'yf_json/'+s+'.json','a')
        fp.write(js)
        fp.close()
        
        sector = ticker.info['sector']
        longname = ticker.info['longName']
        industry = ticker.info['industry']
        
        ticker_sector_dict[s] = sector
        ticker_industry_dict[s] = industry
        ticker_longname_dict[s] = longname
        
    except:
        ticker_sector_dict[s] = "NA"
        ticker_industry_dict[s] = "NA"
        ticker_longname_dict[s] = "NA"      
```

```python
lenrow, _ = df.shape
sectors = list(np.zeros(lenrow))
industries = list(np.zeros(lenrow))
longnames = list(np.zeros(lenrow))
```

```python
for i in range(lenrow):
    if i%10000==0:
        print(i)
    symbol = df.loc[i]['Symbol']
    sectors[i] = ticker_sector_dict[symbol]
    industries[i] = ticker_industry_dict[symbol]
    longnames[i] = ticker_longname_dict[symbol]
```

```python
df['Sector'] = sectors
df['Industry'] = industries
df['LongName'] = longnames
```
