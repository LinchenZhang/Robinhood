```python
from pytrends.request import TrendReq
import pandas as pd
import time
import os
import glob

base = "C:/Users/Linchen Zhang/Desktop/fall2020/linchen_robintrack/"
indir = "C:/Users/Linchen Zhang/Desktop/fall2020/linchen_robintrack/robintrack-popularity-history/tmp/popularity_export"
clean = "C:/Users/Linchen Zhang/Desktop/fall2020/linchen_robintrack/cleanPopularity/"

startTime = time.time()
```

```python
pytrend = TrendReq()
df = pd.read_csv(base + "with_sec_ind_long.csv")
df.drop(df[df['LongName'] == None].index, inplace = True) 
```

```python
df
```

```python
dataset = []
longnames = df['LongName'].unique()
longnames.shape
# longnames = ['APPLE Inc.']
```

```python
a = list(longnames)
a[2] = str(a[2])
longnames = a
longnames
```

```python
count = 0
for longname in longnames:
    count += 1
    if count%100 == 0:
        print(count)
    keywords = [longname]
    pytrend.build_payload(kw_list=keywords,timeframe='2017-01-01 2020-09-01')
    data = pytrend.interest_over_time()

    if not data.empty:
        data = data.drop(labels=['isPartial'],axis='columns')
        dataset.append(data)

result = pd.concat(dataset, axis=1)
result.to_csv('google_trends.csv')
```
